```python
# 1. split the comma-separated columns into lists (guarded so re-running is safe)
df['genres'] = df['genres'].apply(lambda x: x.split(', ') if isinstance(x, str) else x)
df['cast']   = df['cast'].apply(lambda x: x.split(', ')[:3] if isinstance(x, str) else x)  # top 3 actors

# 2. make description & director the same "list" shape as the others
df['description'] = df['description'].apply(lambda x: x.split() if isinstance(x, str) else x)
df['director']    = df['director'].apply(lambda x: [x] if isinstance(x, str) else x)

# 3. remove spaces INSIDE names (the "Tom Hanks" trap) so tokens are unique
df['cast']     = df['cast'].apply(lambda x: [i.replace(' ', '') for i in x])
df['director'] = df['director'].apply(lambda x: [i.replace(' ', '') for i in x])
df['genres']   = df['genres'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
df['tags'] = df['description'] + df['genres'] + df['cast'] + df['director']
```

```python
df['tags'] = df['tags'].apply(lambda x: ' '.join(x).lower())
```

```python
# tags for matching; metadata for ranking/filtering later
new_df = df[['show_id', 'title', 'tags', 'rating']]
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
extras = new_df['title'].str.contains(
    r'Directing|Behind the|Making of|Featurette|: The Making|Bonus', case=False, na=False)
print("Removing", extras.sum(), "extras")
new_df = new_df[~extras].reset_index(drop=True)
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
vectors = tfidf.fit_transform(new_df['tags']).toarray()
```

```python
try:
    feature_names = cv.get_feature_names_out()
except AttributeError:
    feature_names = cv.get_feature_names()
feature_names[:200]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
index = new_df[new_df['title'] == 'Interstellar'].index[0]
sorted(similarity[index], reverse=True)[1:6]
```

```python
def recommend(movie):
    matches = new_df[new_df['title'] == movie]
    if matches.empty:
        print(f"'{movie}' not found. Check the spelling!")
        return

    index = matches.index[0]
    distances = similarity[index]

    # Stage 1: pool of similar movies
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:21]

    candidates = []
    for i in movie_list:
        candidates.append({
            'title': new_df.iloc[i[0]].title,
            'similarity': round(i[1], 3),
            'rating': new_df.iloc[i[0]].rating
        })
    candidates_df = pd.DataFrame(candidates)

    # Stage 2: BLEND both signals into one score
    candidates_df['score'] = (
        candidates_df['similarity'] * 0.85 +          # relevance = 70%
        (candidates_df['rating'] / 10) * 0.15          # quality   = 30%
    )
    return candidates_df.sort_values('score', ascending=False).head(5)

    
```

```python
recommend('Inception')
```

```python
print("Top 5 recommendations for 'Interstellar':")
recommend('Interstellar')
```
